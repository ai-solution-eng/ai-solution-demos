## S3 Interaction

```python
import boto3
from botocore.exceptions import ClientError
import os
```

## List buckets and upload file

```python
def list_s3_buckets():
    """
    List all buckets in the S3 store
    """
    s3_client = boto3.client('s3')
    try:
        response = s3_client.list_buckets()
        print("Existing Buckets:")
        for bucket in response['Buckets']:
            print(f'  - {bucket["Name"]}')
        for key in s3_client.list_objects(Bucket='testbucket')['Contents']:
            print(key['Key'])
        return response['Buckets']

    except ClientError as e:
        print(f"Error when trying to list buckets: {e}")
        return None

def upload_to_s3(file_path, bucket_name, object_name=None):
    """
    Upload a file into the S3 bucket.

    :param file_path: Path of file to upload.
    :param bucket_name: name of target bucket.
    :param object_name: name of S3 object, if none it's using the file name.
    :return: True for success, otherwise False.
    """
    if object_name is None:
        object_name = os.path.basename(file_path)

    s3_client = boto3.client('s3')
    try:
        #Create bucket if it's not existing yet
        s3_client.create_bucket(Bucket=bucket_name)
    except ClientError as e:
        error_code = e.response['Error']['Code']
        if error_code == 'BucketAlreadyOwnedByYou':
            print(f"Bucket '{bucket_name}' already exists and is owned by you. Skipping creation.")
        else:
            print(f"An unexpected error occurred: {e}")
    try:
        s3_client.upload_file(file_path, bucket_name, object_name)
        print(f"File '{file_path}' successful uploaded in bucket '{bucket_name}' as '{object_name}' .")
        return True
    except FileNotFoundError:
        print(f"The file under '{file_path}' wasn't found. ")
        return False
    except ClientError as e:
        print(f"Error when trying to upload the file: {e}")
        return False

if __name__ == '__main__':
    # 1. list buckets
    list_s3_buckets()

    print("\n" + "="*30 + "\n")

    # 2. upload file 
    # Exchange this to your file path
    file_to_upload = "/mnt/user/predictive-maintenance-service-o4atcftdbci4cavb.bento"
    target_bucket = "testbucket" # change it to your bucket

    # If file to upload does not exist create an example file
    if not os.path.exists(file_to_upload):
        with open(file_to_upload, "w") as f:
            f.write("This is a testfile for S3 Upload.")

    upload_to_s3(file_to_upload, target_bucket)
```
